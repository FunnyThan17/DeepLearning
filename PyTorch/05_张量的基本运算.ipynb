```python
"""
案例:
    演示张量的基本运算.

涉及到的API:
    add(), sub(), mul(), div(), neg()       -> 加减乘除, 取反, substract, multiply, divide
    add_(), sub_(), mul_(), div_(), neg_()  -> 功能同上, 只不过可以修改源数据, 类似于 Pandas部分的 inplace = True

需要你记忆的:
    1. 可以用 +, -, *, / 符号来替代 上述的 加减乘除函数.
    2. 如果是张量 和 数值运算, 则: 该数值会和张量中的每个值依次进行 对应的运算.
"""
import torch
```

## 1.加法

```python
# 1.创建张量
t1 = torch.tensor([1, 2, 3])
t2 = t1.add(10)     # 不会修好原数据
# t2 = t1 + 10      # 效果同上

t3 = torch.tensor([1, 2, 3])
t4 = t3.add_(10)    # 会修改原数据

print(f't1: {t1}')
print(f't2: {t2}')
print('=' * 50)
print(f't3: {t3}')
print(f't4: {t4}')

```

## 2.减法

```python
# 1.创建张量
t1 = torch.tensor([1, 2, 3])
t2 = t1.sub(10)     # 不会修好原数据
# t2 = t1 - 10      # 效果同上

t3 = torch.tensor([1, 2, 3])
t4 = t3.sub_(10)    # 会修改原数据

print(f't1: {t1}')
print(f't2: {t2}')
print('=' * 50)
print(f't3: {t3}')
print(f't4: {t4}')

```

## 取反

```python
# 1.创建张量
t1 = torch.tensor([1, 2, 3])
t2 = t1.neg()     # 不会修好原数据

t3 = torch.tensor([1, 2, 3])
t4 = t3.neg_()    # 会修改原数据

print(f't1: {t1}')
print(f't2: {t2}')
print('=' * 50)
print(f't3: {t3}')
print(f't4: {t4}')

```
