```python
"""
案例:
    演示张量的点乘 和 矩阵乘法操作.

点乘:
    要求: 两个张量的维度保持一致, 对应元素直接做 乘法操作.
    API:
        t1 * t2
        t1.mul(t2)          # multiply: 乘法

矩阵乘法:
    要求: 两个张量, 第一个张量 的列数, 等于 第二个张量 的行数(A列 = B行)
    结果: A行B列
    API:
        t1 @ t2
        t1.matmul(t2)
        t1.dot(t2)          扩展: 只针对于一维张量有效.
"""

import torch
```

```python
# 1.定义张量,2行3列
t1 = torch.tensor([[2, 3, 4], [5, 6, 7]])
print(f't1: {t1}')

t2 = torch.tensor([[1, 3, 5], [2, 4, 6]])
print(f't2: {t2}')

# 2.点乘
t3 = t1 * t2
print(f't3: {t3}')
t4 = t1.mul(t2)
print(f't4: {t4}')

# 3. 矩阵相乘, 要求: A列 = B行
t5 = torch.tensor([[2, 3, 4], [5, 6, 7], [8, 9, 10]])
t6 = torch.tensor([[1, 3, 5], [2, 4, 6], [7, 8, 9]])

t7 = t5 @ t6
t8 = t5.matmul(t6)
print(f't7: {t7}')
print(f't8: {t8}')
```
